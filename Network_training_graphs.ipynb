# Network results 

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Classic, 50 epochs, 20 Networks

```python
aTest0 = np.load('classic_50eps_20nets_test.npy')
aTrain0 = np.load('classic_50eps_20nets_train.npy')
```

```python
n_epochs0 = 50
```

```python
plt.plot(range(n_epochs0), aTest0, 'r')
plt.plot(range(n_epochs0), aTrain0, 'b')
plt.yticks(np.arange(0.88, 1.0, 0.01))
plt.grid(b=True)
plt.legend(['Test', 'Train'])
#plt.figure(figsize=(5,10)), only really shows up when saving
```
